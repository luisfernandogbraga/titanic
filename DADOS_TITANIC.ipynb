Análise de Dados: Visualização de Dados

```python
import seaborn as sns
```

```python
data= sns.load_dataset('titanic')
display(data)
```

```python
data.info()
```

```python
# valor da passagem por sobrevivente
x= data[['fare', 'survived']]
x= x.query('fare < 100') # quero só os que pagaram menos que 100

with sns.axes_style('whitegrid'):
    grafico= sns.histplot(data=x ,x='fare', hue='survived')
    grafico.set(title='Valor da passagem por sobrevivência', xlabel='Valor (USD)', ylabel='Contagem');
    grafico.get_legend().set_title('Sobreviveu ?')
```

O gráfico mostra que o preço das passagems infuenciou na sobrevivência das vítimas, quem pagou menos, certamente morreu

O **KDE** (*kernel density estimator*), assim como o **histograma**, representa a distribuição de uma variável numérica, mas em uma distribuição contínua. A variável numérica é segmentada em intervalos representados por uma função contínua estimada. Ajuda a evidênciar a distribuição da variável numérica.

```python
x= data[['age', 'survived']]

with sns.axes_style('whitegrid'):
    grafico= sns.histplot(data= x, x='age', hue='survived', kde=True)
    grafico.set(title='Idade por sobrevivência', xlabel='Idade', ylabel='Contagem');
    grafico.get_legend().set_title('Sobreviveu ?');
```

O gráfico mostra que a idade não enfluênciou em nada sobre a sobrevivência.

```python
titanic = data[["age", "survived", "class"]]

with sns.axes_style('whitegrid'):
  
  grafico = sns.FacetGrid(data=titanic, col="class",  hue="survived") # agora estamos coluna 
  grafico.map(sns.histplot, "age", kde=True)
  print('As ')
```

```python
with sns.axes_style('whitegrid'):

  grafico = sns.boxplot(x=data["age"])
  grafico.set(title='Distribuição de Idade', xlabel='Idade');
```

Os dados estão concentrados entre 20 e aproximadamente 40 anos, sendo que tenham tinham crianças de 1 anos há idosos de mais de 60 anos de idade.

```python
with sns.axes_style('whitegrid'):

  grafico = sns.boxplot(x=data["age"], y=data["class"])
  grafico.set(title="Distribuição de idade por classe", xlabel="Idade", ylabel="Classe");
```

Os dados nos mostram que, na primeira classe, nós temos passageiros de 1 ano aproximadamente até os 80 anos, sendo que a idade concentrada dos passageiros está entre 29 a 50 anos de idade. Já na segunda classe temos passageiros com aproximadamente 2 anos a 55 anos de idade, Sendo a sua concentração em entre 25 anos de idade até os 45 anos. Já na terceira classe, nós temos renascidos até passageiros com 50 anos de idade, e a sua base se concentra entre os 18 anos até os 33 anos de idade aproximadamente.
